```python
import pandas as pd
```

```python
df = pd.read_csv("maestro-v3.0.0.csv")
```

```python
df = df[df['year'] == 2018]
```

```python
midi_names = df['midi_filename']
```

```python
melody_names = [n.replace(".midi", "-melody.midi") for n in midi_names]
melody_names = [n.replace("2018","2018-melody") for n in melody_names]
```

```python
df['melody_filename'] = melody_names
```

```python
df.to_csv("metadata.csv",index=False)
```
