```python
import pandas as pd

df=pd.read_csv("results.csv")
df.rename(columns={'recall1': 'Recall@1',
                   'recall3': 'Recall@3',
                   'recall5':"Recall@5",
                   'recall10':"Recall@10",
                   "mean_rank":"Mean Rank",
                   "norm_sim":"MNS(Mean Normalised Similarity)",
                   "margin":"MD(Margin of Discrimination)",
                   "avg_confidence":"Mean Confidence"
                  },
          inplace=True, errors='raise')
df
```

```python
recall=df[["Similarity Method","Melody Extractor","Processed","Recall@1","Recall@3","Recall@5","Recall@10"]]
recall.to_csv("recall.csv",index=False)
recall
```

```python
MR=df[["Similarity Method","Melody Extractor","Processed","Mean Rank"]]
MR.to_csv("mean_rank.csv",index=False)
MR
```

```python
MNS=df[["Similarity Method","Melody Extractor","Processed","MNS(Mean Normalised Similarity)"]]
MNS.to_csv("mns.csv",index=False)
MNS
```
