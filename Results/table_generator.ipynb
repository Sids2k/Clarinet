```python
import pandas as pd

df=pd.read_csv("results.csv")
df.rename(columns={'recall1': 'Recall@1',
                   'recall3': 'Recall@3',
                   'recall5':"Recall@5",
                   'recall10':"Recall@10",
                   "mean_rank":"Mean Rank",
                   "norm_sim":"MNS(Mean Normalised Similarity)",
                   "margin":"MD(Margin of Discrimination)",
                   "avg_confidence":"Mean Confidence",
                   'mrr': "Mean Reciprocal Rank (MRR)"
                  },
          inplace=True, errors='raise')
df
```

```python
recall=df[["Similarity Method","Melody Extractor","Processed","Recall@1","Recall@3","Recall@5","Recall@10"]]
recall.to_csv("recall.csv",index=False)
recall
```

```python
sim_list = list(recall['Similarity Method'][:4])
melody_list = list(recall['Melody Extractor'][:4])
recall_1 = list(recall['Recall@1'][:4])
recall_3 = list(recall['Recall@3'][:4])
recall_5 = list(recall['Recall@5'][:4])
recall_10 = list(recall['Recall@10'][:4])
```

```python
labels = []
for i in range(len(sim_list)):
    labels.append(sim_list[i] + '-' + melody_list[i])
```

```python
rec_list = [recall_1, recall_3, recall_5, recall_10]
for i,p in enumerate([1,3,5,10]):
    barplot =  plt.bar(labels, rec_list[i], label=f'Recall@{p}',width=0.4)
    barplot[0].set_color('red')
    barplot[1].set_color('magenta')
    barplot[2].set_color('blue')
    barplot[3].set_color('green')
    plt.xlabel('Method')
    plt.ylabel(f'Recall@{p}')
    plt.title(f'Recall@{p}')
    plt.savefig(f'recall_{p}.png',dpi=1200)
```

```python
allscores = []
for i in range(4):
    allscores.append([recall_1[i], recall_3[i], recall_5[i], recall_10[i]])
```

```python
plt.plot([1,3,5,10],allscores[0],label = 'Text-Modified',color='red')
plt.plot([1,3,5,10],allscores[1],label = 'Text-Skyline',color='red')
plt.plot([1,3,5,10],allscores[2],label = 'Time-Modified',color = 'blue')
plt.plot([1,3,5,10],allscores[3],label = 'Time-Skyline',color='green')
plt.xlabel('Recall@k',fontsize=14)
plt.ylabel('k',fontsize=14)
plt.title('Recall@k',fontsize=14)
plt.legend()
plt.savefig('recall_all.png',dpi=1200)
```

```python
import json
with open('timeres.json') as f:
    timeres = json.load(f)
```

```python
for key, value in timeres.items():
    timeres[key] = value/40
```

```python
# convert timeres to dataframe
timeres_df = pd.DataFrame.from_dict(timeres, orient='index')
# drop those keys which contain 'processed'
timeres_df = timeres_df.drop(timeres_df[timeres_df.index.str.contains('processed')].index)
```

```python
timeres_df.rename(columns={0:'Time'},inplace=True)
```

```python
# sort by time
timeres_df = timeres_df.sort_values(by='Time')
```

```python
timeres_df.to_csv('timeres.csv',index=False)
```
