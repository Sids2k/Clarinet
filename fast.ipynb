```python
from Clarinet.utils.fast import fast
from evaluate import computeScores
import os
```

```python
collection_dir="Text/Original Collection"

query_folders_location="Text/Noisy Queries"


inputs=[] # Tupples of input params ie. 

for query_dir in os.listdir(query_folders_location):

    output_dir=f"Results/NoisyResults/{query_dir}"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    query_dir=f"{query_folders_location}/{query_dir}"
    inputs.append((query_dir,collection_dir,-1,-1,0,"text",output_dir))
fast(computeScores,inputs,num_processes=4)
```
